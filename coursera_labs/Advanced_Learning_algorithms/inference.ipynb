### Building a neural network architecture

```python
model = tf.keras.Sequential([
    tf.keras.layers.Dense(units=25, activation='sigmoid'),
    tf.keras.layers.Dense(units=15, activation='sigmoid'),
    tf.keras.layers.Dense(units=1, activation='sigmoid')
])
x = np.array([[200.0, 17.0]])
y = np.array([[1, 0]])
x_new = np.array([[150.0, 20.0]])
model.compile(optimizer='sgd', loss='mean_squared_error')
model.fit(x, y, epochs=1000)
print(model.predict(x_new))
```

## Forward prop in Python

1. Hardcoding for every single version

```python
# layer 1
x = np.array([[200.0, 17.0]])
w1_1 = np.array([1, 2])
b1_1 = np.array([-1])
z1_1 = np.dot(x, w1_1) + b1_1
a1_1 = 1 / (1 + np.exp(-z1_1))


w1_2 = np.array([3, 1])
b1_2 = np.array([-1])
z1_2 = np.dot(x, w1_2) + b1_2
a1_2 = 1 / (1 + np.exp(-z1_2))


w1_3 = np.array([1, 1])
b1_3 = np.array([-1])
z1_3 = np.dot(x, w1_3) + b1_3
a1_3 = 1 / (1 + np.exp(-z1_3))

a1 = np.array([a1_1, a1_2, a1_3])


# Layer 2
w2_1 = np.array([7, 1, 1])
b2_1 = np.array([-1])
z2_1 = np.dot(a1, w2_1) + b2_1
a2_1 = 1 / (1 + np.exp(-z2_1))
```

2. General implementation of forward propagation

```python
W = np.array([
    [1, -3, 5],
    [2, -1, -1]
])

b = np.array([2, 1, -1])
a_in = np.array([-2, 4])


def dense(a_in, W, b):
    units = W.shape[1]
    a_out = np.zeros(units)
    for j in range(units):
        w = W[:, j]
        z = np.dot(w, a_in) + b[j]
        a_out[j] = 1 / (1 + np.exp(-z))

    return a_out


def sequential(x):
    a1 = dense(x, W1, b1)
    a2 = dense(a1, W2, b2)
    a3 = dense(a2, W3, b3)
    a4 = dense(a3, W4, b4)
    f_x = a4
    return f_x
```
